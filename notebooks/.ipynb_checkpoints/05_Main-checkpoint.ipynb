```python
import os
import re
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import EngFormatter, MaxNLocator

# --- Constants (can be overridden in function calls) ---
SNIPER_RESULTS_DIR = "/home/skataoka26/COSC_498/miniMXE/results/sniper_llc_32mb_16w_20251002T020050Z"
DYNAMORIO_LOGS_DIR = "/home/skataoka26/COSC_498/miniMXE/results_trace/logs"
TIMESTAMP_PREFIX = "20250929T203551Z"

# --- Core Helper Functions (Unchanged) ---
def safe_div(numer, denom):
    numer = np.array(numer, dtype=float)
    denom = np.array(denom, dtype=float)
    with np.errstate(divide='ignore', invalid='ignore'):
        out = np.where((denom > 0) & np.isfinite(denom), numer / denom, np.nan)
    return out

def geomean(values):
    vals = np.array([v for v in np.ravel(values) if v > 0 and np.isfinite(v)], dtype=float)
    return float(np.exp(np.mean(np.log(vals)))) if len(vals) else np.nan

def parse_sniper(base_dir):
    """
    Walk output_* folders, read summary.csv and energy_bounds.csv,
    merge them on (benchmark, config). Return a tidy DataFrame.
    """
    energy_rows = []
    summary_rows = []
    print(f"[parse_sniper] Searching under: {base_dir}")
    for root, _, files in os.walk(base_dir):
        if os.path.basename(root).startswith("output_") and "summary.csv" in files and "energy_bounds.csv" in files:
            energy_path = os.path.join(root, "energy_bounds.csv")
            summary_path = os.path.join(root, "summary.csv")
            try:
                e = pd.read_csv(energy_path)
                s = pd.read_csv(summary_path)
            except Exception:
                continue
            e_keep = ["benchmark", "config", "time_s", "energy_exact_J", "leak_J", "dyn_exact_nJ"]
            e = e[[c for c in e_keep if c in e.columns]].copy()
            s_base = ["benchmark", "config", "ipc", "time_ns", "l3_miss_rate_pct"]
            keep_s = [c for c in s_base if c in s.columns]
            for col in s.columns:
                cl = col.lower()
                if ("l3" in cl or "llc" in cl) and (
                    "access" in cl or "read" in cl or "write" in cl or "hit" in cl or "miss" in cl or "evict" in cl or "wb" in cl
                ):
                    if col not in keep_s:
                        keep_s.append(col)
            s = s[keep_s].copy()
            for col in ["time_s", "energy_exact_J", "leak_J", "dyn_exact_nJ"]:
                if col in e.columns:
                    e[col] = pd.to_numeric(e[col], errors="coerce")
            for col in s.columns:
                if col not in ("benchmark", "config"):
                    s[col] = pd.to_numeric(s[col], errors="coerce")
            if "dyn_exact_nJ" in e.columns:
                e["dyn_exact_J"] = e["dyn_exact_nJ"] * 1e-9
            energy_rows.append(e)
            summary_rows.append(s)
    if not energy_rows or not summary_rows:
        raise FileNotFoundError(f"Could not find summary.csv and energy_bounds.csv under: {base_dir}")
    E = pd.concat(energy_rows, ignore_index=True)
    S = pd.concat(summary_rows, ignore_index=True)
    df = pd.merge(E, S, on=["benchmark", "config"], how="inner")
    df = df.dropna(subset=["benchmark", "config", "time_s", "energy_exact_J"])
    print(f"[parse_sniper] Merged DataFrame shape: {df.shape}")
    return df

def _kv_from_line(s):
    kv = {}
    m = re.search(r"\bscope=(\w+)", s)
    if m:
        kv["scope"] = m.group(1)
    for k, v in re.findall(r"([A-Za-z0-9_]+)=([+\-]?(?:\d+(?:\.\d*)?|\.\d+)(?:[eE][+\-]?\d+)?|nan)", s):
        try:
            kv[k] = float("nan") if v.lower() == "nan" else float(v)
        except Exception:
            pass
    return kv

def parse_dynamorio(logs_dir, ts_prefix):
    """
    Parse DynamoRIO log files by aggregating *per-interval* rows correctly.
    File name format: {ts_prefix}_<benchmark>_instr.rwstats.log
    """
    rows = []
    print(f"[parse_dynamorio] Searching logs: {logs_dir} (prefix={ts_prefix})")
    if not os.path.isdir(logs_dir):
        print("[parse_dynamorio] Logs dir not found; returning empty DataFrame.")
        return pd.DataFrame()
    r_weight_keys = {"read_entropy", "read_local_entropy", "read_footprint90L"}
    w_weight_keys = {"write_entropy", "write_local_entropy", "write_footprint90L"}
    cumulative_last_keys = [
        "reads", "writes", "bytes_read", "bytes_written",
        "global_footprint_bytes", "read_unique_lines", "write_unique_lines",
    ]
    max_keys = ["read_unique", "write_unique", "uniq_lines", "uniq_pages", "footprint_bytes"]
    found = 0
    for fname in os.listdir(logs_dir):
        if not (fname.startswith(ts_prefix) and fname.endswith("_instr.rwstats.log")):
            continue
        m = re.match(rf"{re.escape(ts_prefix)}_(.+?)_instr\.rwstats\.log$", fname)
        if not m:
            continue
        found += 1
        benchmark = m.group(1)
        fpath = os.path.join(logs_dir, fname)
        ivals = []
        try:
            with open(fpath, "r", errors="ignore") as f:
                for line in f:
                    if "scope=" not in line:
                        continue
                    kv = _kv_from_line(line.strip())
                    if not kv:
                        continue
                    sc = str(kv.get("scope", "")).lower()
                    if sc in ("interval", "final"):
                        ivals.append(kv)
        except Exception:
            continue
        if not ivals:
            continue
        reads_seq = [r.get("reads", np.nan) for r in ivals]
        writes_seq = [r.get("writes", np.nan) for r in ivals]
        dreads = [np.nan]
        dwrites = [np.nan]
        for i in range(1, len(ivals)):
            a, b = reads_seq[i-1], reads_seq[i]
            c, d = writes_seq[i-1], writes_seq[i]
            dreads.append((b - a) if (np.isfinite(a) and np.isfinite(b) and b >= a) else np.nan)
            dwrites.append((d - c) if (np.isfinite(c) and np.isfinite(d) and d >= c) else np.nan)
        r_wts, w_wts = [], []
        for i, r in enumerate(ivals):
            rtot = r.get("read_total", np.nan)
            wtot = r.get("write_total", np.nan)
            r_wts.append(rtot if np.isfinite(rtot) else dreads[i])
            w_wts.append(wtot if np.isfinite(wtot) else dwrites[i])
        sum_read_total = float(np.nansum([w for w in r_wts if np.isfinite(w)])) if any(np.isfinite(w) for w in r_wts) else np.nan
        sum_write_total = float(np.nansum([w for w in w_wts if np.isfinite(w)])) if any(np.isfinite(w) for w in w_wts) else np.nan
        agg = {"benchmark": benchmark, "scope": "aggregate"}
        agg["read_total"] = sum_read_total
        agg["write_total"] = sum_write_total
        def _wavg(keys, weights):
            den = float(np.nansum([w for w in weights if np.isfinite(w)]))
            for k in keys:
                num = 0.0
                if den > 0:
                    for i, row in enumerate(ivals):
                        v = row.get(k, np.nan)
                        w = weights[i]
                        if np.isfinite(v) and np.isfinite(w) and w > 0:
                            num += v * w
                    agg[k] = num / den if den > 0 else np.nan
                else:
                    agg[k] = np.nan
        _wavg(r_weight_keys, r_wts)
        _wavg(w_weight_keys, w_wts)
        for k in cumulative_last_keys:
            for row in reversed(ivals):
                v = row.get(k, np.nan)
                if np.isfinite(v):
                    agg[k] = v
                    break
        if np.isfinite(agg.get("read_unique_lines", np.nan)):
            agg["read_unique"] = agg["read_unique_lines"]
        else:
            rmax = np.nan
            for row in ivals:
                v = row.get("read_unique", np.nan)
                if np.isfinite(v):
                    rmax = v if not np.isfinite(rmax) else max(rmax, v)
            if np.isfinite(rmax):
                agg["read_unique"] = rmax
        if np.isfinite(agg.get("write_unique_lines", np.nan)):
            agg["write_unique"] = agg["write_unique_lines"]
        else:
            wmax = np.nan
            for row in ivals:
                v = row.get("write_unique", np.nan)
                if np.isfinite(v):
                    wmax = v if not np.isfinite(wmax) else max(wmax, v)
            if np.isfinite(wmax):
                agg["write_unique"] = wmax
        for k in max_keys:
            vmax = np.nan
            for row in ivals:
                v = row.get(k, np.nan)
                if np.isfinite(v):
                    vmax = v if not np.isfinite(vmax) else max(vmax, v)
            if np.isfinite(vmax):
                agg[k] = vmax
        for k in ("instrs", "instructions"):
            s = float(np.nansum([row.get(k, np.nan) for row in ivals if np.isfinite(row.get(k, np.nan))]))
            lastv = np.nan
            for row in reversed(ivals):
                v = row.get(k, np.nan)
                if np.isfinite(v):
                    lastv = v
                    break
            if np.isfinite(s) or np.isfinite(lastv):
                agg[k] = max(s if np.isfinite(s) else -np.inf, lastv if np.isfinite(lastv) else -np.inf)
        rows.append(agg)
    print(f"[parse_dynamorio] Processed {found} files; rows={len(rows)}")
    return pd.DataFrame(rows)

def process_dynamorio_data(dr):
    """Processes raw DynamoRIO data to generate a feature DataFrame."""
    if dr.empty:
        warnings.warn("DynamoRIO data not found or empty.")
        return pd.DataFrame()
    feat_map = {
        "rtotal": "read_total", "runique": "read_unique", "90%f_tr": "read_footprint90L",
        "Hrg": "read_entropy", "Hrl": "read_local_entropy", "wtotal": "write_total",
        "wunique": "write_unique", "90%f_tw": "write_footprint90L", "Hwg": "write_entropy",
        "Hwl": "write_local_entropy",
    }
    feats = pd.DataFrame({"benchmark": dr["benchmark"]})
    for out_name, src in feat_map.items():
        feats[out_name] = pd.to_numeric(dr.get(src), errors="coerce")
    feats = feats.set_index("benchmark")
    feats['read_intensity'] = safe_div(feats.get('runique'), feats.get('rtotal'))
    feats['write_intensity'] = safe_div(feats.get('wunique'), feats.get('wtotal'))
    feats['rw_ratio_total'] = safe_div(feats.get('rtotal'), feats.get('wtotal'))
    feats['rw_ratio_unique'] = safe_div(feats.get('runique'), feats.get('wunique'))
    feats['rw_ratio_global_entropy'] = safe_div(feats.get('Hrg'), feats.get('Hwg'))
    feats['rw_ratio_local_entropy'] = safe_div(feats.get('Hrl'), feats.get('Hwl'))
    feats['rw_ratio_90_footprint'] = safe_div(feats.get('90%f_tr'), feats.get('90%f_tw'))
    return feats

# --- Plotting Helper Functions (Unchanged) ---
def corr_heatmap(ax, matrix, row_labels, col_labels, title):
    im = ax.imshow(matrix, aspect="auto", vmin=-1, vmax=1, cmap="coolwarm")
    ax.set_yticks(np.arange(len(row_labels)))
    ax.set_yticklabels(row_labels)
    ax.set_xticks(np.arange(len(col_labels)))
    ax.set_xticklabels(col_labels, rotation=45, ha="right")
    ax.set_title(title, fontsize=11)
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            val = matrix[i, j]
            if np.isfinite(val):
                ax.text(j, i, f"{val:.2f}", ha="center", va="center",
                        color="white" if abs(val) > 0.6 else "black", fontsize=8)
    return im

def plot_grouped_feature_bars_v2(feats, benchmarks, feature_cols, norm="p95", clip=(0, 1.25),
                                 yscale="linear", title=None, out_file=None):
    bms = [b for b in benchmarks if b in feats.index]
    cols = [c for c in feature_cols if c in feats.columns and not feats[c].isna().all()]
    if not bms or not cols:
        raise ValueError("No matching benchmarks or features to plot.")
    df = feats.loc[bms, cols].astype(float).copy()
    if norm == "p95":
        for c in df.columns:
            p95 = np.nanpercentile(df[c].values, 95)
            denom = p95 if (np.isfinite(p95) and p95 > 0) else 1.0
            df[c] /= denom
        ylbl = "value / P95 (per feature)"
    elif norm == "none":
        ylbl = "value"
    else:
        raise ValueError(f"Unknown norm: {norm}")
    if clip is not None:
        df = df.clip(lower=clip[0], upper=clip[1])
    n_bm, n_feat = len(df.index), len(df.columns)
    x = np.arange(n_bm)
    width = min(0.80 / max(n_feat, 1), 0.18)
    fig_w = max(10, 0.6 * n_bm + 4)
    fig, ax = plt.subplots(1, 1, figsize=(fig_w, 5.2))
    for i, c in enumerate(df.columns):
        offsets = (i - (n_feat - 1) / 2.0) * width
        y = np.nan_to_num(df[c].values, nan=0.0)
        ax.bar(x + offsets, y, width=width, label=c)
    ax.set_xticks(x)
    ax.set_xticklabels(df.index.tolist(), rotation=45, ha="right")
    ax.set_ylabel(ylbl)
    ax.set_title(title or f"Features per Benchmark — norm={norm}")
    ax.grid(axis="y", alpha=0.25)
    ax.set_axisbelow(True)
    if yscale == "log":
        ax.set_yscale("log")
    if n_feat > 6:
        ax.legend(loc="upper left", bbox_to_anchor=(1.01, 1.0), title="feature", frameon=False)
        fig.tight_layout(rect=[0, 0, 0.85, 1])
    else:
        ax.legend(ncol=min(n_feat, 6), frameon=False)
        fig.tight_layout()
    if out_file:
        fig.savefig(out_file, dpi=200, bbox_inches="tight")
        print(f"Saved plot to {out_file}")
    return fig, ax, df

def plot_feature_corr_pair(feats, feature_cols, title="Feature ↔ Feature Correlation", out_file=None):
    cols = [c for c in feature_cols if c in feats.columns and not feats[c].isna().all()]
    if len(cols) < 2:
        raise ValueError("Need at least two feature columns for correlation.")
    M = feats[cols].astype(float)
    cS = M.corr(method="spearman")
    cP = M.corr(method="pearson")
    size = max(6, 0.5 * len(cols) + 4)
    fig, axes = plt.subplots(1, 2, figsize=(2 * size + 2, size))
    im0 = corr_heatmap(axes[0], cS.values, cols, cols, "Spearman")
    im1 = corr_heatmap(axes[1], cP.values, cols, cols, "Pearson")
    fig.colorbar(im1, ax=axes.ravel().tolist(), fraction=0.035, pad=0.02, label="correlation")
    fig.suptitle(title, y=0.995, fontsize=13)
    fig.tight_layout(rect=[0, 0, 1, 0.96])
    if out_file:
        fig.savefig(out_file, dpi=200, bbox_inches="tight")
        print(f"Saved plot to {out_file}")
    return cS, cP

def plot_feature_facets_bars(feats, benchmarks, feature_cols, ncols=5, nrows=2,
                             title="Features (raw units)", out_file=None):
    label_formatter = EngFormatter(places=3, sep="")
    bms = [b for b in benchmarks if b in feats.index]
    cols = [c for c in feature_cols if c in feats.columns and not feats[c].isna().all()]
    if not bms or not cols:
        raise ValueError("No matching benchmarks or features to plot.")
    w_per = 2.8 + 2.8 + 0.05 * len(bms)
    h_per = 2.6 + 2.6
    fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * w_per, nrows * h_per))
    axes = np.ravel(axes)
    x = np.arange(len(bms))
    axis_formatter = EngFormatter(places=3)
    total_slots = nrows * ncols
    for i in range(total_slots):
        ax = axes[i]
        if i >= len(cols):
            ax.axis("off")
            continue
        c = cols[i]
        y = feats.loc[bms, c].astype(float).values
        finite_y = y[np.isfinite(y)]
        lo = float(np.min(finite_y)) if finite_y.size else 0.0
        hi = float(np.max(finite_y)) if finite_y.size else 1.0
        span = max(hi - lo, 1e-12)
        ax.set_ylim(lo - 0.06 * span, hi + 0.06 * span)
        ax.bar(x, np.nan_to_num(y, nan=0.0), width=0.8)
        ax.set_title(c, fontsize=9)
        ax.set_xticks(x)
        ax.set_xticklabels(bms, rotation=45, ha="right", fontsize=7)
        ax.yaxis.set_major_locator(MaxNLocator(nbins=4, prune='both'))
        ax.yaxis.set_major_formatter(axis_formatter)
        ax.tick_params(axis="y", labelsize=7)
        ax.grid(axis="y", alpha=0.25)
        ax.set_axisbelow(True)
        y_lo, y_hi = ax.get_ylim()
        offset = (y_hi - y_lo) * 0.02
        for xi, yi in enumerate(y):
            if np.isfinite(yi):
                ax.text(xi, yi + offset, label_formatter(yi), ha="center", va="bottom", fontsize=7)
    fig.suptitle(title, y=0.995, fontsize=12)
    fig.tight_layout(rect=[0, 0, 1, 0.96])
    if out_file:
        fig.savefig(out_file, dpi=200, bbox_inches="tight")
        print(f"Saved plot to {out_file}")
    return fig, axes

# ==============================================================================
# --- NEW MODULAR FUNCTIONS ---
# ==============================================================================

def load_and_process_data(runid=None, ts_prefix=None, logs_dir=None, out_dir=None):
    """
    Loads, parses, and processes Sniper and DynamoRIO data.
    This is the main data preparation step. It returns a dictionary
    of all the necessary DataFrames and lists for subsequent analysis.
    """
    if out_dir is None: out_dir = os.getcwd()
    os.makedirs(out_dir, exist_ok=True)

    # --- Resolve Paths ---
    def _valid_sniper_dir(p):
        return os.path.isdir(p) and any(os.path.basename(r).startswith("output_") and "summary.csv" in f for r, _, f in os.walk(p))

    if runid:
        candidate = runid if os.path.isabs(runid) else os.path.join(os.path.dirname(SNIPER_RESULTS_DIR), runid)
        sniper_dir = candidate if _valid_sniper_dir(candidate) else SNIPER_RESULTS_DIR
    else:
        sniper_dir = SNIPER_RESULTS_DIR
    
    logs_dir = logs_dir or DYNAMORIO_LOGS_DIR
    ts_prefix = ts_prefix or TIMESTAMP_PREFIX
    
    print(f"[dataproc] Using SNIPER dir: {sniper_dir}")
    print(f"[dataproc] Using DynamoRIO logs: {logs_dir} (prefix={ts_prefix})")
    
    # --- Load and Process ---
    sniper_raw = parse_sniper(sniper_dir)
    dynamorio_raw = parse_dynamorio(logs_dir, ts_prefix)
    dynamorio_features = process_dynamorio_data(dynamorio_raw)

    sniper_raw.to_csv(os.path.join(out_dir, "sniper_raw.csv"), index=False)
    if not dynamorio_raw.empty:
        dynamorio_raw.to_csv(os.path.join(out_dir, "dynamorio_raw.csv"), index=False)

    # --- Create Targets and Join ---
    sram = sniper_raw[sniper_raw["config"] == "SRAM"].set_index("benchmark")
    jans = sniper_raw[sniper_raw["config"] == "JanS"].set_index("benchmark")
    common = sram.index.intersection(jans.index)
    if len(common) < 2: raise RuntimeError("Need at least 2 common benchmarks for correlations.")
    
    benchmark_order = sram.loc[common].sort_values("time_s").index.tolist()
    
    targets_abs = jans.loc[benchmark_order, ["dyn_exact_J", "leak_J", "energy_exact_J", "time_s"]].rename(columns={
        "dyn_exact_J": "Dyn_energy", "leak_J": "Leakage",
        "energy_exact_J": "LLC_energy", "time_s": "exe_time"})
    
    ratio_targets = pd.DataFrame(index=benchmark_order)
    ratio_targets["dyn_energy_ratio"] = safe_div(jans.loc[benchmark_order, "dyn_exact_J"], sram.loc[benchmark_order, "dyn_exact_J"])
    ratio_targets["energy_ratio"] = safe_div(jans.loc[benchmark_order, "energy_exact_J"], sram.loc[benchmark_order, "energy_exact_J"])
    ratio_targets["exe_time_ratio"] = safe_div(jans.loc[benchmark_order, "time_s"], sram.loc[benchmark_order, "time_s"])
    
    corr_df = dynamorio_features.join(targets_abs, how="inner").join(ratio_targets, how="inner")
    corr_df = corr_df.reindex(benchmark_order).dropna(axis=0, how='all')
    
    corr_df.to_csv(os.path.join(out_dir, "corr_input_joined.csv"))

    # --- Identify Feature and Target Columns ---
    candidate_features = [
        "rtotal","runique","90%f_tr","Hrg","Hrl", "wtotal","wunique","90%f_tw",
        "Hwg","Hwl", "read_intensity","write_intensity", "rw_ratio_total",
        "rw_ratio_unique", "rw_ratio_global_entropy","rw_ratio_local_entropy",
        "rw_ratio_90_footprint"
    ]
    feature_names = [f for f in candidate_features if f in corr_df.columns and not corr_df[f].isna().all()]
    target_candidates = ["Dyn_energy", "Leakage", "LLC_energy", "exe_time",
                         "dyn_energy_ratio", "energy_ratio", "exe_time_ratio"]
    target_names = [c for c in target_candidates if c in corr_df.columns and not corr_df[c].isna().all()]
    
    print(f"[dataproc] Final DataFrame for correlation has shape: {corr_df.shape}")
    
    return {
        "corr_df": corr_df,
        "dynamorio_features": dynamorio_features,
        "feature_names": feature_names,
        "target_names": target_names,
        "benchmark_order": benchmark_order
    }

def plot_feature_target_correlation(data_package, out_dir=None):
    """Plots the heatmap of memory features vs. performance/energy targets."""
    corr_df = data_package["corr_df"]
    feature_names = data_package["feature_names"]
    target_names = data_package["target_names"]
    
    corr_mat = corr_df[feature_names + target_names].corr(method="pearson").loc[feature_names, target_names]
    
    fig, ax = plt.subplots(1, 1, figsize=(10, max(4, 0.30 * len(feature_names))))
    im = corr_heatmap(ax, corr_mat.values, feature_names, target_names,
                      "Features vs JanS (absolute + ratios) — Pearson’s r")
    fig.colorbar(im, ax=ax, fraction=0.045, pad=0.02, label="pearson correlation")
    fig.tight_layout(rect=[0, 0, 1, 0.96])
    
    if out_dir:
        out_file = os.path.join(out_dir, "corr_heatmap_pearson.png")
        fig.savefig(out_file, dpi=200, bbox_inches="tight")
        print(f"Saved plot to {out_file}")
    plt.show()

def plot_normalized_feature_bars(data_package, out_dir=None):
    """Generates P95-normalized grouped bar plots for top benchmarks."""
    feats = data_package["dynamorio_features"]
    benchmarks = data_package["benchmark_order"]
    top_benchmarks = benchmarks[:12]

    feature_cols = ["rtotal","wtotal","runique","wunique","90%f_tr","90%f_tw",
                    "Hrg","Hrl","Hwg","Hwl","read_intensity","write_intensity"]
    present_cols = [c for c in feature_cols if c in feats.columns and not feats[c].isna().all()]
    
    out_file = os.path.join(out_dir, "features_grouped_bar_normalised_p95.png") if out_dir else None
    plot_grouped_feature_bars_v2(
        feats, benchmarks=top_benchmarks, feature_cols=present_cols, norm="p95",
        clip=(0, 1.25), yscale="linear", 
        title=f"DynamoRIO features — P95-normalised ({len(top_benchmarks)} benchmarks)",
        out_file=out_file
    )
    plt.show()

def plot_raw_log_feature_bars(data_package, out_dir=None):
    """Generates raw value, log-scale grouped bar plots for top benchmarks."""
    feats = data_package["dynamorio_features"]
    benchmarks = data_package["benchmark_order"]
    top_benchmarks = benchmarks[:12]
    
    feature_cols = ["rtotal","wtotal","runique","wunique","90%f_tr","90%f_tw",
                    "Hrg","Hrl","Hwg","Hwl","read_intensity","write_intensity"]
    present_cols = [c for c in feature_cols if c in feats.columns and not feats[c].isna().all()]

    out_file = os.path.join(out_dir, "features_grouped_bar_raw_logy.png") if out_dir else None
    plot_grouped_feature_bars_v2(
        feats, benchmarks=top_benchmarks, feature_cols=present_cols, norm="none",
        clip=None, yscale="log",
        title=f"DynamoRIO features — RAW (log y) ({len(top_benchmarks)} benchmarks)",
        out_file=out_file
    )
    plt.show()

def plot_inter_feature_correlations(data_package, out_dir=None):
    """Plots the correlation matrix (Spearman & Pearson) between DynamoRIO features."""
    feats = data_package["dynamorio_features"]
    feature_cols = ["rtotal","wtotal","runique","wunique","90%f_tr","90%f_tw",
                    "Hrg","Hrl","Hwg","Hwl","read_intensity","write_intensity"]
    present_cols = [c for c in feature_cols if c in feats.columns and not feats[c].isna().all()]

    out_file = os.path.join(out_dir, "features_vs_features_corr_pair.png") if out_dir else None
    if len(present_cols) >= 2:
        plot_feature_corr_pair(feats, feature_cols=present_cols, out_file=out_file)
        plt.show()

def plot_feature_facets_bars(feats, benchmarks, feature_cols, ncols=5, nrows=2,
                             title="Features (raw units)", out_file=None):
    label_formatter = EngFormatter(places=3, sep="")
    bms = [b for b in benchmarks if b in feats.index]
    cols = [c for c in feature_cols if c in feats.columns and not feats[c].isna().all()]
    if not bms or not cols:
        raise ValueError("No matching benchmarks or features to plot.")
    
    w_per = 2.8 + 2.8 + 0.05 * len(bms)
    h_per = 2.6 + 2.6
    fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * w_per, nrows * h_per))
    axes = np.ravel(axes)
    x = np.arange(len(bms))
    axis_formatter = EngFormatter(places=3)
    total_slots = nrows * ncols

    for i in range(total_slots):
        ax = axes[i]
        if i >= len(cols):
            ax.axis("off")
            continue
        
        c = cols[i]
        y = feats.loc[bms, c].astype(float).values

        # --- ADD THIS BLOCK TO PRINT THE VALUES ---
        print(f"\n✅ Values for subplot: '{c}'")
        print("-" * 30)
        for benchmark_name, value in zip(bms, y):
            if np.isfinite(value):
                # Using the same formatter as the plot labels for consistency
                print(f"  - {benchmark_name:<15}: {label_formatter(value)}")
            else:
                print(f"  - {benchmark_name:<15}: NaN")
        # --- END OF ADDED BLOCK ---

        finite_y = y[np.isfinite(y)]
        lo = float(np.min(finite_y)) if finite_y.size else 0.0
        hi = float(np.max(finite_y)) if finite_y.size else 1.0
        span = max(hi - lo, 1e-12)
        
        ax.set_ylim(lo - 0.06 * span, hi + 0.06 * span)
        ax.bar(x, np.nan_to_num(y, nan=0.0), width=0.8)
        ax.set_title(c, fontsize=9)
        ax.set_xticks(x)
        ax.set_xticklabels(bms, rotation=45, ha="right", fontsize=7)
        ax.yaxis.set_major_locator(MaxNLocator(nbins=4, prune='both'))
        ax.yaxis.set_major_formatter(axis_formatter)
        ax.tick_params(axis="y", labelsize=7)
        ax.grid(axis="y", alpha=0.25)
        ax.set_axisbelow(True)
        
        y_lo, y_hi = ax.get_ylim()
        offset = (y_hi - y_lo) * 0.02
        
        for xi, yi in enumerate(y):
            if np.isfinite(yi):
                ax.text(xi, yi + offset, label_formatter(yi), ha="center", va="bottom", fontsize=7)
                
    fig.suptitle(title, y=0.995, fontsize=12)
    fig.tight_layout(rect=[0, 0, 1, 0.96])
    
    if out_file:
        fig.savefig(out_file, dpi=200, bbox_inches="tight")
        print(f"Saved plot to {out_file}")
        
    return fig, axes
        
# ==============================================================================
# --- EXTENSION: COMBINED 2MB vs 32MB ANALYSIS ---
# ==============================================================================

# --- Visual and Physical Constants for Combined Plots ---
CLR_SRAM_S = "#ADD8E6"  # light blue (SRAM, Small LLC)
CLR_JANS_S = "#90EE90"  # light green (JanS, Small LLC)
CLR_SRAM_L = "#6FA8DC"  # darker blue (SRAM, Large LLC)
CLR_JANS_L = "#5CBF7B"  # darker green (JanS, Large LLC)

HATCH_LEAK = "//"
HATCH_DYN  = ".."
HATCH_RH   = "///"
HATCH_WH   = "\\\\"
HATCH_MS   = "xx"

# Per-access energies (in picojoules) from NVSim
E_PJ = {
    "2MB": {
        "SRAM": {"read": 562.0,  "write": 534.0,  "miss": 562.0},
        "JanS": {"read": 859.0,  "write": 489.0,  "miss": 859.0},
    },
    "32MB": {
        "SRAM": {"read": 1650.0, "write": 1486.0, "miss": 1650.0},
        "JanS": {"read": 1458.0, "write": 1107.0, "miss": 1458.0},
    },
}
pJ2J = 1e-12

# --- Internal Helpers for Combined Plotting ---

def _pivot_by_config(df, cols):
    """Pivots a DataFrame to have configs as columns."""
    keep = ["benchmark","config"] + [c for c in cols if c in df.columns]
    pv = df[keep].pivot_table(index="benchmark", columns="config", values=[c for c in cols if c in df.columns])
    pv.columns = [f"{a}_{b}" for a,b in pv.columns]
    return pv

def _load_and_pivot_sniper_results(sniper_dir):
    """Loads and processes a single sniper result directory for combined analysis."""
    df = parse_sniper(sniper_dir)
    df = df[df["config"].isin(["SRAM","JanS"])].copy()
    
    # Pivot core metrics
    base_cols = ["time_s", "energy_exact_J", "leak_J", "dyn_exact_nJ"]
    base = _pivot_by_config(df, base_cols)
    if "dyn_exact_nJ_SRAM" in base.columns: base["dyn_exact_J_SRAM"] = pd.to_numeric(base["dyn_exact_nJ_SRAM"], errors="coerce") * 1e-9
    if "dyn_exact_nJ_JanS" in base.columns: base["dyn_exact_J_JanS"] = pd.to_numeric(base["dyn_exact_nJ_JanS"], errors="coerce") * 1e-9

    # Pivot L3 counters
    l3_cols = [c for c in df.columns if c.lower().startswith("l3_")]
    cnt = _pivot_by_config(df, l3_cols) if l3_cols else pd.DataFrame(index=base.index)
    out = base.join(cnt, how="left")

    # Derive missing miss rates
    for dev in ["SRAM", "JanS"]:
        rate, ms, acc = f"l3_miss_rate_pct_{dev}", f"l3_misses_db_{dev}", f"l3_accesses_db_{dev}"
        if rate not in out and all(c in out for c in [ms, acc]):
            out[rate] = safe_div(pd.to_numeric(out[ms]), pd.to_numeric(out[acc])) * 100.0
    return out

def _group4(ax, labels, s_sram, s_jans, l_sram, l_jans, title, ylabel):
    """Helper to plot four grouped bars per benchmark."""
    n = len(labels); x = np.arange(n); w = 0.18
    ax.bar(x - 3*w/2, s_sram, width=w, color=CLR_SRAM_S, label="SRAM (S)")
    ax.bar(x -   w/2, s_jans, width=w, color=CLR_JANS_S, label="JanS (S)")
    ax.bar(x +   w/2, l_sram, width=w, color=CLR_SRAM_L, label="SRAM (L)")
    ax.bar(x + 3*w/2, l_jans, width=w, color=CLR_JANS_L, label="JanS (L)")
    ax.set_xticks(x); ax.set_xticklabels(labels, rotation=45, ha="right")
    ax.set_title(title); ax.set_ylabel(ylabel)
    ax.legend(ncol=4, frameon=False, loc='upper center', bbox_to_anchor=(0.5, -0.25))
    ax.grid(axis='y', linestyle='--', alpha=0.7)

def _stacked4(ax, labels, s_sram_parts, s_jans_parts, l_sram_parts, l_jans_parts, comp_order, comp_hatch, title, ylabel):
    """Helper to plot four STACKED bars per benchmark."""
    n = len(labels); x = np.arange(n); w = 0.18
    bottoms = {k: np.zeros(n) for k in ["S_SRAM","S_JANS","L_SRAM","L_JANS"]}
    colors  = {"S_SRAM": CLR_SRAM_S, "S_JANS": CLR_JANS_S, "L_SRAM": CLR_SRAM_L, "L_JANS": CLR_JANS_L}
    xpos    = {"S_SRAM": x - 3*w/2,  "S_JANS": x - w/2,    "L_SRAM": x + w/2,    "L_JANS": x + 3*w/2}
    data_map = {"S_SRAM": s_sram_parts, "S_JANS": s_jans_parts, "L_SRAM": l_sram_parts, "L_JANS": l_jans_parts}

    for comp in comp_order:
        for key in xpos.keys():
            values = np.nan_to_num(data_map[key][comp])
            ax.bar(xpos[key], values, width=w, color=colors[key], hatch=comp_hatch[comp], edgecolor="k", bottom=bottoms[key])
            bottoms[key] += values

    ax.set_xticks(x); ax.set_xticklabels(labels, rotation=45, ha="right")
    ax.set_title(title); ax.set_ylabel(ylabel)
    devs = [plt.Rectangle((0,0),1,1,color=c,label=l) for c,l in zip([CLR_SRAM_S, CLR_JANS_S, CLR_SRAM_L, CLR_JANS_L],["SRAM (S)","JanS (S)","SRAM (L)","JanS (L)"])]
    comps = [plt.Rectangle((0,0),1,1,facecolor="white",edgecolor="k",hatch=h,label=l) for l,h in comp_hatch.items()]
    ax.legend(handles=devs+comps, ncol=4, frameon=False, loc='upper center', bbox_to_anchor=(0.5, -0.25))
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# --- New Data Loading Function for Combined Analysis ---

def load_combined_data(dir_small_llc, dir_large_llc):
    """Loads and aligns Sniper data from two directories (e.g., 2MB and 32MB LLCs)."""
    print("--- Loading and aligning data for combined analysis ---")
    df_small = _load_and_pivot_sniper_results(dir_small_llc)
    df_large = _load_and_pivot_sniper_results(dir_large_llc)
    
    common_benchmarks = df_small.index.intersection(df_large.index)
    df_small = df_small.loc[common_benchmarks]
    df_large = df_large.loc[common_benchmarks]
    
    # Order benchmarks by the execution time of SRAM in the small LLC case
    order = df_small.sort_values("time_s_SRAM").index.tolist()
    
    print(f"Found {len(order)} common benchmarks. Ordered by SRAM (Small LLC) execution time.")
    
    return {
        "df_small": df_small.loc[order],
        "df_large": df_large.loc[order],
        "benchmark_order": order
    }

# --- New Plotting Functions for Combined Analysis ---

def plot_combined_timing(combined_data, plot_type='raw', out_dir=None):
    """Plots combined timing, either as 'raw' seconds or 'ratio' (JanS/SRAM)."""
    df_s = combined_data["df_small"]
    df_l = combined_data["df_large"]
    order = combined_data["benchmark_order"]
    
    fig, ax = plt.subplots(figsize=(max(14, 0.95 * len(order)), 6))
    
    if plot_type == 'raw':
        _group4(ax, order, df_s["time_s_SRAM"], df_s["time_s_JanS"], df_l["time_s_SRAM"], df_l["time_s_JanS"],
                "Execution Time (raw)", "Time (s)")
        fname = "bar_time_raw_combined.png"
    elif plot_type == 'ratio':
        _group4(ax, order, np.ones(len(order)), safe_div(df_s["time_s_JanS"], df_s["time_s_SRAM"]),
                np.ones(len(order)), safe_div(df_l["time_s_JanS"], df_l["time_s_SRAM"]),
                "Execution Time (ratio JanS/SRAM)", "Ratio")
        fname = "bar_time_ratio_combined.png"
    else:
        raise ValueError("plot_type must be 'raw' or 'ratio'")

    fig.tight_layout(rect=[0, 0.1, 1, 1])
    if out_dir:
        out_file = os.path.join(out_dir, fname)
        fig.savefig(out_file, dpi=200, bbox_inches="tight")
        print(f"Saved plot to {out_file}")
    plt.show()

def plot_combined_total_energy(combined_data, plot_type='raw', out_dir=None):
    """Plots total LLC energy, either 'raw' (stacked) or 'ratio'."""
    df_s = combined_data["df_small"]
    df_l = combined_data["df_large"]
    order = combined_data["benchmark_order"]
    
    fig, ax = plt.subplots(figsize=(max(14, 1.05 * len(order)), 6.5))

    if plot_type == 'raw':
        parts_s_s = {"Leakage": df_s["leak_J_SRAM"], "Dynamic": df_s["dyn_exact_J_SRAM"]}
        parts_s_j = {"Leakage": df_s["leak_J_JanS"], "Dynamic": df_s["dyn_exact_J_JanS"]}
        parts_l_s = {"Leakage": df_l["leak_J_SRAM"], "Dynamic": df_l["dyn_exact_J_SRAM"]}
        parts_l_j = {"Leakage": df_l["leak_J_JanS"], "Dynamic": df_l["dyn_exact_J_JanS"]}
        _stacked4(ax, order, parts_s_s, parts_s_j, parts_l_s, parts_l_j,
                  ["Leakage","Dynamic"], {"Leakage": HATCH_LEAK, "Dynamic": HATCH_DYN},
                  "Total LLC Energy (raw, breakdown)", "Energy (J)")
        fname = "bar_energy_total_raw_combined.png"
    elif plot_type == 'ratio':
        _group4(ax, order,
                np.ones(len(order)), safe_div(df_s["energy_exact_J_JanS"], df_s["energy_exact_J_SRAM"]),
                np.ones(len(order)), safe_div(df_l["energy_exact_J_JanS"], df_l["energy_exact_J_SRAM"]),
                "Total LLC Energy (ratio JanS/SRAM)", "Ratio")
        fname = "bar_energy_total_ratio_combined.png"
    else:
        raise ValueError("plot_type must be 'raw' or 'ratio'")
        
    fig.tight_layout(rect=[0, 0.1, 1, 1])
    if out_dir:
        out_file = os.path.join(out_dir, fname)
        fig.savefig(out_file, dpi=200, bbox_inches="tight")
        print(f"Saved plot to {out_file}")
    plt.show()

def plot_combined_dynamic_energy(combined_data, plot_type='raw', out_dir=None):
    """Plots dynamic LLC energy, either 'raw' (stacked by op) or 'ratio'."""
    df_s = combined_data["df_small"]
    df_l = combined_data["df_large"]
    order = combined_data["benchmark_order"]
    
    fig, ax = plt.subplots(figsize=(max(14, 1.05 * len(order)), 6.5))
    
    if plot_type == 'raw':
        def get_breakdown(df, cap_label):
            ep = E_PJ[cap_label]
            sram = {"Read hits": df[f"l3_read_hits_SRAM"] * ep["SRAM"]["read"]  * pJ2J,
                    "Write hits": df[f"l3_write_hits_SRAM"] * ep["SRAM"]["write"] * pJ2J,
                    "Misses":     df[f"l3_misses_db_SRAM"] * ep["SRAM"]["miss"]  * pJ2J}
            jans = {"Read hits": df[f"l3_read_hits_JanS"] * ep["JanS"]["read"]  * pJ2J,
                    "Write hits": df[f"l3_write_hits_JanS"] * ep["JanS"]["write"] * pJ2J,
                    "Misses":     df[f"l3_misses_db_JanS"] * ep["JanS"]["miss"]  * pJ2J}
            return sram, jans
        s_s, s_j = get_breakdown(df_s, "2MB")
        l_s, l_j = get_breakdown(df_l, "32MB")
        
        _stacked4(ax, order, s_s, s_j, l_s, l_j,
                  ["Read hits","Write hits","Misses"], {"Read hits": HATCH_RH, "Write hits": HATCH_WH, "Misses": HATCH_MS},
                  "Dynamic Energy (raw, RH/WH/Miss)", "Energy (J)")
        fname = "bar_energy_dyn_raw_combined.png"
    elif plot_type == 'ratio':
        _group4(ax, order,
                np.ones(len(order)), safe_div(df_s["dyn_exact_J_JanS"], df_s["dyn_exact_J_SRAM"]),
                np.ones(len(order)), safe_div(df_l["dyn_exact_J_JanS"], df_l["dyn_exact_J_SRAM"]),
                "Dynamic Energy (ratio JanS/SRAM)", "Ratio")
        fname = "bar_energy_dyn_ratio_combined.png"
    else:
        raise ValueError("plot_type must be 'raw' or 'ratio'")

    fig.tight_layout(rect=[0, 0.1, 1, 1])
    if out_dir:
        out_file = os.path.join(out_dir, fname)
        fig.savefig(out_file, dpi=200, bbox_inches="tight")
        print(f"Saved plot to {out_file}")
    plt.show()

def plot_combined_miss_rate(combined_data, plot_type='raw', out_dir=None):
    """Plots L3 miss rate, either as 'raw' percentage or 'ratio' (JanS/SRAM)."""
    df_s = combined_data["df_small"]
    df_l = combined_data["df_large"]
    order = combined_data["benchmark_order"]
    
    fig, ax = plt.subplots(figsize=(max(14, 0.95 * len(order)), 6))

    mr_s_s = df_s["l3_miss_rate_pct_SRAM"]
    mr_s_j = df_s["l3_miss_rate_pct_JanS"]
    mr_l_s = df_l["l3_miss_rate_pct_SRAM"]
    mr_l_j = df_l["l3_miss_rate_pct_JanS"]
    
    if plot_type == 'raw':
        _group4(ax, order, mr_s_s, mr_s_j, mr_l_s, mr_l_j, "L3 Miss Rate (raw)", "Miss rate (%)")
        fname = "bar_missrate_raw_combined.png"
    elif plot_type == 'ratio':
        _group4(ax, order,
                np.ones(len(order)), safe_div(mr_s_j, mr_s_s),
                np.ones(len(order)), safe_div(mr_l_j, mr_l_s),
                "L3 Miss Rate (ratio JanS/SRAM)", "Ratio")
        fname = "bar_missrate_ratio_combined.png"
    else:
        raise ValueError("plot_type must be 'raw' or 'ratio'")
    
    fig.tight_layout(rect=[0, 0.1, 1, 1])
    if out_dir:
        out_file = os.path.join(out_dir, fname)
        fig.savefig(out_file, dpi=200, bbox_inches="tight")
        print(f"Saved plot to {out_file}")
    plt.show()

def run_full_analysis(runid=None, out_dir="."):
    """
    Convenience function to run the full pipeline: loads data and generates all plots.
    """
    print("--- 1. Loading and Processing Data ---")
    data = load_and_process_data(runid=runid, out_dir=out_dir)
    
    print("\n--- 2. Plotting Feature vs. Target Correlation ---")
    plot_feature_target_correlation(data, out_dir=out_dir)
    
    print("\n--- 3. Plotting Normalized Feature Bars ---")
    plot_normalized_feature_bars(data, out_dir=out_dir)
    
    print("\n--- 4. Plotting Raw Feature Bars (Log Scale) ---")
    plot_raw_log_feature_bars(data, out_dir=out_dir)
    
    print("\n--- 5. Plotting Inter-Feature Correlations ---")
    plot_inter_feature_correlations(data, out_dir=out_dir)

    print("\n--- 6. Plotting Faceted Feature Bars ---")
    plot_faceted_feature_bars(data, out_dir=out_dir)
    
    print(f"\nAnalysis complete. Artifacts saved to: {out_dir}")
```
